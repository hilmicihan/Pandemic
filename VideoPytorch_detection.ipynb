```python
from facenet_pytorch import MTCNN
import torch
import numpy as np
import mmcv, cv2
from PIL import Image, ImageDraw
from IPython import display
```

```python

##device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
##print('Running on device: {}'.format(device))
```

```python
mtcnn = MTCNN(keep_all=True)
```

```python
path_data = "data/"
video_name1 = "misircarsi.mp4"
video_name= path_data+video_name1

video = mmcv.VideoReader(video_name)
frames = [Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)) for frame in video]

frames[::5]
```

```python
frames_tracked = []
for i, frame in enumerate(frames):
    ##print('\rTracking frame: {}'.format(i + 1), end='')
    
    # Detect faces
    boxes, probs, points = mtcnn.detect(frame,landmarks=True)   
    # Draw faces
    frame_draw = frame.copy()
    draw = ImageDraw.Draw(frame_draw)
    try:
        for box in boxes:
            draw.rectangle(box.tolist(), outline=(255, 0, 0), width=1)
            ## frame_draw.show()
            draw = ImageDraw.Draw(frame_draw)
        frames_tracked.append(frame_draw)
    except:
        continue
```

```python
path_result = "data/video_results/"+video_name1

dim = frames_tracked[0].size
fourcc = cv2.VideoWriter_fourcc(*'FMP4')    
video_tracked = cv2.VideoWriter(path_result, fourcc, 25.0, dim)
for frame in frames_tracked:
    video_tracked.write(cv2.cvtColor(np.array(frame), cv2.COLOR_RGB2BGR))
video_tracked.release()
```
